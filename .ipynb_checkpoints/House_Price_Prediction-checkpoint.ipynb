## Pre Processing of size is done

# Pre Processing of total_sqft is to be done now 

```python
data['total_sqft'].unique()
```

Creating a function to handle values separated with " - "

```python
def clean_mean(sqft):
    tokens=sqft.split('-')
    if len(tokens)==2:
        return (float(tokens[0]) + float(tokens[1]))/2

    else :
        try : 
            return float(sqft)
        except:
            return None
        
```

```python
data['total_sqft']=data['total_sqft'].apply(lambda x:clean_mean(x))
```

```python
data
```

```python
data['total_sqft'].unique()
```

```python
data.describe()
```

As we can see that there are less count for total_sqft as there few none or null values we will remove it !

```python
data.dropna(inplace=True)
```

```python
data
```

### Till now total_sqft is cleaned, we will now make few more columns like sqft per bed or price per sqft.
Also we will consider only those data which are not absurd or make sense for ex we will remove if there is sqft per  bed is 0 100 or below 300.

```python

data['total_sqft'] = pd.to_numeric(data['total_sqft'])
data['bedrooms'] = pd.to_numeric(data['bedrooms'])


data['sqft_per_bed'] = data['total_sqft'] / data['bedrooms']
```

```python
data
```

```python
data.sqft_per_bed.describe()
```

### Creating a new data set

```python
data2= data[data['sqft_per_bed']>=300]
```

```python
data2['price_per_sqft']=data2['price']*100000/data2['total_sqft']
```

### Now we categorical value in 'location' columns rest all are the numerical value columns, so we will do the encoding of it and model training .

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),
    remainder='passthrough',
    verbose_feature_names_out=False,
    force_int_remainder_cols=False
)
```

```python
lr=LinearRegression()
```

```python
scaler=StandardScaler()
```

```python
model=make_pipeline(col_trans,scaler,lr)
```

```python
data_input=data3.drop(columns=['price'])
data_output=data3['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_input,data_output,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input=pd.DataFrame([['Electronic City Phase II',1500,3.0,2.0,2]],columns=['location','total_sqft','bath','balcony','bedrooms'])
```

```python
model.predict(input)
```
